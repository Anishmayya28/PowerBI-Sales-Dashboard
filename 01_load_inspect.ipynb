```python
import sqlite3
import pandas as pd

conn = sqlite3.connect('mydb.db')
cursor = conn.cursor()
```

```python
# FACT Table
cursor.execute('''
CREATE TABLE IF NOT EXISTS fact_sales (
    order_id TEXT,
    product_id TEXT,
    customer_id TEXT,
    order_date DATE,
    region TEXT,
    sales FLOAT,
    quantity INTEGER,
    discount FLOAT,
    profit FLOAT
)
''')

# DIMENSION Tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_customer (
    customer_id TEXT PRIMARY KEY,
    customer_name TEXT,
    segment TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_product (
    product_id TEXT PRIMARY KEY,
    product_name TEXT,
    category TEXT,
    sub_category TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_order_date (
    order_date DATE PRIMARY KEY,
    year INTEGER,
    month INTEGER,
    day INTEGER,
    weekday TEXT
)
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS dim_region (
    region TEXT PRIMARY KEY,
    state TEXT,
    city TEXT
)
''')

conn.commit()
```

```python
df = pd.read_csv('superstore.csv', encoding='ISO-8859-1')
df.head()
```

```python
df.to_sql('raw_sales', conn, index=False, if_exists='replace')
```

```python
cursor.execute('''
INSERT OR IGNORE INTO dim_customer (customer_id, customer_name, segment)
SELECT DISTINCT [Customer ID], [Customer Name], [Segment] FROM raw_sales
''')
```

```python
cursor.execute('''
INSERT OR IGNORE INTO dim_product (product_id, product_name, category, sub_category)
SELECT DISTINCT [Product ID], [Product Name], [Category], [Sub-Category] FROM raw_sales
''')
```

```python
cursor.execute('''
INSERT OR IGNORE INTO dim_order_date (order_date, year, month, day, weekday)
SELECT DISTINCT 
    [Order Date],
    CAST(STRFTIME('%Y', [Order Date]) AS INTEGER),
    CAST(STRFTIME('%m', [Order Date]) AS INTEGER),
    CAST(STRFTIME('%d', [Order Date]) AS INTEGER),
    STRFTIME('%w', [Order Date])
FROM raw_sales
''')
```

```python
cursor.execute('''
INSERT INTO fact_sales (order_id, product_id, customer_id, order_date, region, sales, quantity, discount, profit)
SELECT 
    [Order ID],
    [Product ID],
    [Customer ID],
    [Order Date],
    [Region],
    [Sales],
    [Quantity],
    [Discount],
    [Profit]
FROM raw_sales
''')
```

```python
conn.commit()

# Verify number of rows in fact_sales
pd.read_sql_query("SELECT COUNT(*) FROM fact_sales", conn)
```
